```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_base = pd.read_csv("raw_dataset_from_scrapping.csv")

display(df_base)

df_base.info()
```

```python
# Tratamento da tabela

# Renomeando coluna
df_analise = df_base.rename(columns={"Unnamed: 0": "Id"})

# Adicionando nova coluna
df_analise["Preço"] = df_analise["Reais"] + (df_analise["Centavos"]/100)

#Mudando de posição a coluna da tabela
coluna = df_analise.pop("Preço")
df_analise.insert(2,"Preço",coluna)
df_analise = df_analise.drop(columns=["Reais","Centavos"])

# Removendo caracteres indesejaveis da coluna N_Avaliações

df_analise["N_Avaliações"] = (
    df_analise["N_Avaliações"]
    .str.replace("(","")
    .str.replace(")","")
 )

# Tratando valores nulos e dataType 
df_analise["N_Avaliações"].value_counts()

df_analise["N_Avaliações"] = df_analise["N_Avaliações"].fillna(0)

df_analise["N_Avaliações"] = df_analise["N_Avaliações"].astype("int64")


# Tirando caracteres indesejaveis da coluna
df_analise["Desconto"].value_counts()

df_analise["Desconto"] = (
    df_analise["Desconto"]
    .str.replace("%","")
    .str.replace(" ", "")
    .str.replace("OFF","")
 )

# Mudando dataType, tratando valores nulos e tratando valores inteiro para decimal
df_analise["Desconto"] = df_analise["Desconto"].astype("float64")

df_analise["Desconto"] = df_analise["Desconto"] / 100

df_analise["Desconto"] = df_analise["Desconto"].fillna(0.0)

# Trazendo o valor original do produto
valor_desconto = df_analise["Preço"] * df_analise["Desconto"]

df_analise["Preço"] = (df_analise["Preço"] + valor_desconto).round(2)

# Criação de nova coluna
df_analise["Preco_com_desconto"] = (df_analise["Preço"] * (1 - df_analise["Desconto"])).round(2)

# Ordenando posições das colunas na tabela novamente
coluna2 = df_analise.pop("Desconto")
df_analise.insert(3,"Desconto",coluna2)

coluna3 = df_analise.pop("Preco_com_desconto")
df_analise.insert(4,"Preco_com_desconto",coluna3)

display(df_analise)

df_analise.info()

```

```python
# Continuação do Tratamento da tabela

# Tabelas com nulos indevidamente : marca, material, genero, temporada

df_analise["Marca"] = df_analise["Marca"].fillna("Desconhecida")

df_analise["Material"] = df_analise["Material"].fillna("Não Informado")

df_analise["Temporada"] = df_analise["Temporada"].fillna("Não Informado")

df_analise["Gênero"] = df_analise["Gênero"].fillna("Não Informado")
df_analise.info()

```

```python
# Analises 

#Produtos com maior preço
df_analise.sort_values(by="Preço",ascending=False)

# Produtos com desconto maior que 30%
df_analise[df_analise["Desconto"] > 0.30]

#Produtos sem avaliação
df_analise[df_analise["N_Avaliações"] == 0]

#Quantidade de produtos por genero
df_analise["Gênero"].value_counts()
```

```python
# Criação de tabela Estatisticas
estatisticas = pd.DataFrame({
        "Estatisticas":["Média","Mediana","Moda","Desvio_Padrão","Valor_maximo","valor_minimo","Amplitude","skew","Kurtosis"],
        "Valor": [
            df_analise["Preço"].mean(),
            df_analise["Preço"].median(),
            df_analise["Preço"].mode()[0],
            df_analise["Preço"].std(),
            df_analise["Preço"].max(),
            df_analise["Preço"].min(),
            df_analise["Preço"].max() -  df_analise["Preço"].min(),
            df_analise["Preço"].skew(),
            df_analise["Preço"].kurt()

        ]
}).round(2).T
display(estatisticas)
```
